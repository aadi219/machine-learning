## Clustering on the california housing dataset from scikit-learn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing
from sklearn.metrics import silhouette_score, silhouette_samples
```

### Data

```python
data = fetch_california_housing(as_frame=True)
```

```python
df = data.frame
print(df)
```

```python
print("Description")
print(data.DESCR)
```

```python
latitudes = df["Latitude"]
longitudes = df["Longitude"]
```

```python
plt.plot(latitudes, longitudes, '.', markersize=1)
plt.xlabel("Latitudes")
plt.ylabel("Longitudes")
plt.show()
```

```python
plt.hist(df.MedHouseVal, 80)
plt.title("Distribution of Median House Values(target) values across the dataset")
plt.show()
```

The dataset shows a clipping of median house values past the value of 500,000

### Preprocessing (Feature Scaling)

```python
from sklearn.preprocessing import StandardScaler
```

```python
# features to be considered for clustering: MedIncome, HouseAge, AveRooms, AveBedrooms, MedHouseVal
filtered_df = df[["MedInc","HouseAge","AveRooms","AveBedrms","MedHouseVal"]]
filtered_df.head()
```

```python
scaled = StandardScaler().fit_transform(filtered_df)
scaled_df = pd.DataFrame(data=scaled, columns=filtered_df.columns)
scaled_df.head()
```

### Clustering (K-Means)

```python
from sklearn.cluster import KMeans
```

```python
inertia = []
for i in range(2, 10):
    kmeans = KMeans(n_clusters=i, init="k-means++",
                    n_init=15, max_iter=500, tol=0.0001,
                    random_state=10, algorithm='lloyd').fit(scaled_df)
    avg_sil_score = silhouette_score(scaled_df, kmeans.labels_)
    print(f"For n={i}, the average silhouette score is = {avg_sil_score}")
    inertia.append(kmeans.inertia_)
```

```python
plt.figure(1, figsize=(14,8))
plt.title("Error vs K")
plt.xlabel("Value of K")
plt.ylabel("Inertia")
plt.plot(np.arange(2,10), inertia, '-')
plt.plot(np.arange(2,10), inertia, '.')
plt.show()
```

In this case, 3 or 4 clusters seems appropraite

#### Analysing the Clusters

```python
kmeans = KMeans(n_clusters=4, init="k-means++",
                    n_init=15, max_iter=500, tol=0.0001,
                    random_state=10, algorithm='lloyd').fit(scaled_df)
clustered_df = filtered_df.copy(True)
labels = kmeans.labels_
clustered_df.insert(5, "Cluster", labels)
print(clustered_df)
```

```python
sil_samples = silhouette_samples(scaled_df, labels)
avg_sil_score = silhouette_score(scaled_df, labels)
y_lower = 10
plt.figure(figsize=(14,10))
for i in range(4):
    cluster_sil_vals = sil_samples[labels==i]
    cluster_sil_vals.sort()
    cluster_size = cluster_sil_vals.shape[0]
    y_upper = y_lower + cluster_size
    plt.fill_between(
        np.arange(y_lower, y_upper), 0,
        cluster_sil_vals
    )
    plt.text(0.5 * cluster_size + y_lower, -0.02, i)
    y_lower = y_upper + 10
plt.axhline(y=avg_sil_score, color="red", linestyle="--")
#plt.yticks([])  # Clear the yaxis labels / ticks
#plt.xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])
plt.show()
```

```python
def get_stats_for_label(label):
    global labels, clustered_df
    print("Label:",label)
    if label not in labels:
        print("Label does not exist.")
        return
    samples = clustered_df.loc[clustered_df.Cluster==label]
    print("Number of samples assigned to label =",  samples.shape[0])
    print("Mean Median House value for label =", np.mean(samples.MedHouseVal))
    print("Standard Deviation of Median House value for label =", np.std(samples.MedHouseVal))
    print("Data")
    print(samples)
```

```python
for label in np.unique(labels):
    get_stats_for_label(label)
```

```python
plt.bar(clustered_df.Cluster.value_counts().index, clustered_df.Cluster.value_counts())
plt.grid(True, axis='y', alpha=0.3)
plt.xlabel("Cluster Label")
plt.title("Distribution of samples across labels")
plt.show()
```

```python
def get_groups(df, labels, k):
    groups = []
    for i in range(k):
        group = df[df.Cluster == i]
        groups.append(group)
    return groups

groups = get_groups(clustered_df, kmeans.labels_, 4)
```

```python
for group in range(len(groups)):
    plt.figure(figsize=(8,4))
    plt.hist(df.MedHouseVal,100)
    plt.hist(groups[group].MedHouseVal, 100)
    plt.title(f"Distribution of Median House Values in cluster {group}\n compared to the overall distribution across the dataset")
    plt.show()
```

W.R.T the Median Value of the Houses, Cluster 0 performed fairly well being able to group many of the high valued houses wihtout many outliers. Cluster 1 and 2 seem to have split themselves between the low-mid range houses with a decent amount of outliers and there seems to be no significant clustering w.r.t MedHouseVal in Cluster 3

```python
for group in range(len(groups)):
    plt.figure(figsize=(8,4))
    plt.hist(df.MedInc,100)
    plt.hist(groups[group].MedInc, 100)
    plt.title(f"Distribution of Median Household Income in cluster {group}\n compared to the overall distribution across the dataset")
    plt.show()
```

Cluster 0 again seems to have done well to group most samples with a high median income together, successfully capturing almost every household with a high income however with some outliers present on lower income side.

```python
colors = ['r','g','b','y']
for i in range(len(groups)):
    plt.plot(groups[i].MedInc, groups[i].MedHouseVal, '.', color=colors[i], markersize=1)
plt.xlabel("Median Income")
plt.ylabel("Median House Value")
plt.legend([f"Cluster {i}" for i in range(len(groups))])
plt.show()
```

This graph shows the clearer distinction between cluster 0 (red) and cluster 1 and 2 (green and blue). However, given the immense overlap between the blue and the green points, it does not show the basis on which cluster 1 and 2 were separated.

```python
plt.plot(df.Latitude, df.Longitude,'.', markersize=1)
plt.plot(df.loc[groups[0].index].Latitude, df.loc[groups[0].index].Longitude, '.', markersize=1, color=colors[0])
plt.title("Location of High Income-High Value houses")
plt.legend(["All houses", "Houses in Cluster 0"])
plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.show()
```

Even without taking location into consideration while clustering, the graph above shows that High Income-High Value households which were grouped together in Cluster 0 tend to be located in similar areas. These households seem to be found closer to the coast and tend to be located closer together giving an idea of the economic status of neighbourhood or districts.

```python
for i in range(len(groups)):
    plt.plot(groups[i].MedInc, df.loc[groups[i].index].AveOccup, '.', color=colors[i], markersize=1)
plt.xlabel("Median Income")
plt.ylabel("Average number of People living in household")
plt.ylim(bottom=0,top=10)
plt.legend([f"Cluster {i}" for i in range(len(groups))])
plt.show()
```

## DBSCAN

Looking at the earlier graphs, it seems as though DBSCAN will not perform significantly well as there are no regions of low density separating high density clusters for DBSCAN to identify.

```python
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
```

```python
# Initial values for hyperparameters
min_pts = 10 # 2 * dim
# Estimating a good value for epsilon - average distance to 9th (min_pts - 1) nearest neighbor
nbrs = NearestNeighbors(n_neighbors=min_pts-1)
nbrs.fit(scaled_df)
dist, _ = nbrs.kneighbors(scaled_df)
```

```python
# dist now contains distances for each point to it's 9 nearest neighbors
dist.shape
```

```python
# considering the distance to only the last neighbor at index ((min_pts - 1) - 1)
distances = np.array([dist[i][min_pts-2] for i in range(len(dist))])
distances.sort()
```

```python
plt.figure(figsize=(12,10))
plt.plot(np.arange(0, distances.size), distances[::-1], '-')
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.plot(np.arange(0, distances.size), distances[::-1], '-')
plt.xticks(np.arange(0, 100, step=5))
plt.yticks(np.linspace(distances.min(), distances.max(), 20))
plt.xlim(left=0, right=100)
plt.show()
```

Elbow of the graph is visible around 3.5 to 8.5 but these values of episilon are too large for this dataset which only shows singular high density clusters. 

```python
# testing smaller epsilon values from 0.3 to 1.2
for eps in np.linspace(0.3, 1.2, 8):
    clustering = DBSCAN(eps=eps, min_samples=min_pts)
    pred = clustering.fit_predict(scaled_df)
    score = silhouette_score(scaled_df, pred)
    print(f"For eps = {eps:.5f}, Average Silhouette Score is: {score:.5f}")        
```

```python
# selecting a smaller eps value due to large amounts of noise and high density of points
eps = 0.4285
dbscan_clustering = DBSCAN(eps=eps, min_samples=min_pts).fit(scaled_df)
dbscan_labels = dbscan_clustering.labels_
print(f"for eps={eps} and min_pts={min_pts}, silhoutte score={silhouette_score(scaled_df, dbscan_labels)}")
```

```python
dbscan_df = filtered_df.copy(True)
dbscan_df.insert(5, "Cluster", dbscan_labels)
dbscan_df.head
```

```python
print(dbscan_df.Cluster.value_counts())
```

Even for a smaller eps=0.428, the DBSCAN algorithm classifies almost all points in cluster 0, some points as noise and barely any points in clusters 1 and 2. This was an expected outcome considering the distribution of points in the dataset.

```python
# amount of noise caught by dbscan
# desirable percentage of noise would be < 30%
n_noise = list(dbscan_labels).count(-1)
print(f"Number of samples categorized as noise: {n_noise}")
print(f"Percentage of noise within the dataset: {(n_noise / dbscan_df.shape[0]) * 100:.4f}%")
```
